```python
outlier_cols = [
    'Year','Life expectancy ','Adult Mortality', 'infant deaths', 'Alcohol', 'percentage expenditure',
    'Hepatitis B', 'Measles ', ' BMI ', 'under-five deaths ', 'Polio',
    'Total expenditure', 'Diphtheria ', ' HIV/AIDS', 'GDP', 'Population',
    ' thinness  1-19 years', ' thinness 5-9 years',
    'Income composition of resources', 'Schooling'
]

for column in outlier_cols:
    plt.figure(figsize=(5, 6))
    plt.boxplot(df[column])
    plt.title(f'Boxplot for {column}')
    plt.ylabel(column)
    plt.show()
```

```python
outlier_cols = [
    'Year','Life expectancy ','Adult Mortality', 'infant deaths', 'Alcohol', 'percentage expenditure',
    'Hepatitis B', 'Measles ', ' BMI ', 'under-five deaths ', 'Polio',
    'Total expenditure', 'Diphtheria ', ' HIV/AIDS', 'GDP', 'Population',
    ' thinness  1-19 years', ' thinness 5-9 years',
    'Income composition of resources', 'Schooling'
]

for column in outlier_cols:

    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 * IQR)
    
   # df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

    '''
    plt.figure(figsize=(5, 4))
    plt.boxplot(df[column])
    plt.show()

    '''
# Calculate the mean of the non-outlier values
mean_value = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)][column].mean()

# Replace outliers with the mean
df[column] = df[column].apply(lambda x: mean_value if x < lower_bound or x > upper_bound else x)
```

```python
x = df.drop(columns=['Life expectancy ']).values
y = df['Life expectancy '].values 
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor

model = KNeighborsRegressor(n_neighbors=3)
model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
x = np.linspace(40,90,1000)
plt.figure(figsize=(10, 6))
plt.scatter(Y_test,Y_pred)
plt.xlabel('Actual Values')
plt.ylabel('Predictions')
plt.plot(x,x,color='black',lw=2.5)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(Y_test, Y_pred)
mse = mean_squared_error(Y_test, Y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(Y_test, Y_pred)

print(mae)
print(mse)
print(rmse)
print(r2)
```
